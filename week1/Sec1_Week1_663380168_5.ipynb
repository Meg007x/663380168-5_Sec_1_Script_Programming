**Lab 1.2 (2) ระกาศตัวแปรและคำนวณ**


```python
num1 = 20
num2 = 21

print(f"{num1} + {num2} ได้ {num1 + num2}") #บวก
print(f"{num1} - {num2} ได้ {num1 - num2}") #ลบ
print(f"{num1} ÷ {num2} ได้ {num1 / num2}") #หาร
print(f"{num1} x {num2} ได้ {num1 * num2}") #คูณ
print(f"{num1} % {num2} ได้ {num1 % num2}") #หารเอาเศษ
print(f"{num1} ^ {num2} ได้ {num1 ** num2}") #ยกกำลัง
```

**Activity 1.2** การแก้โจทย์ปัญหาและปริศนา Python (Python Puzzlers / Solving Math Problems)

```python
#Puzzlers 1
# หาพื้นที่สีเหลี่ยม

width = int(input("กรุณากรอกความกว้าง: "))

length = int(input("กรุณากรอกความยาว: "))

area = width * length

print(f"พื้นที่สี่เหลี่ยมคือ: {area} ตารางเซนติเมตร")
```

```python
#Puzzlers 2
# แปลงอุณหภูมิ
c = int(input("กรุณาใส่อุณหภูมิตอนนี้ : "))
f = (c * 9/5) + 32
print(f"อุณหภูมิตอนนี้คือ: {f} องศาฟาเรนไฮต์")
```

 **คำถาม** เมื่อรันแล้ว z จะมีค่าเท่าไหร่ และ type(z) จะแสดงว่าเป็นข้อมูลชนิดไหน?

```python
#Puzzlers 3
x = 10
y = 3.5
z = x + y
print(z)
print(type(z))
```

**ตอบ** จะได้ z = 13.5 type ของ z คือ float เพราะว่า python แปลงให้เป็น float โดยอัตโนมัติ

**คำถาม** ตัวแปร age ในที่นี้เป็นตัวเลขหรือข้อความ? และผลลัพธ์ของ message จะออกมาเป็นอย่างไร?


```python
#Puzzlers 4
name = "Alice"
age = "30" # สังเกตว่ามีเครื่องหมายคำพูด
message = name + " is " + age + " years old."
print(message)
print(type(age))
```

**ตอบ** age เป็น ข้อความ เพระาว่ามีการใช้ "" ครอบ 30 ไว้ ผลลัพธ์ของ message คือ Alice is 30 years old.

**คำถาม** ผลลัพธ์ของ True and False คืออะไร? และ is_sunny เก็บข้อมูลชนิดไหน?

```python
#Puzzlers 5
is_sunny = True
is_raining = False
print(is_sunny and is_raining)
print(type(is_sunny))
```

**คำตอบ** ผลลัพทธ์ของ True and False ได้ False เพราะ ต้องเป็น T and T ถึงจะได้ค่าเป็น True และ is_sunny เก็บข้อมูลชนิด boolean
