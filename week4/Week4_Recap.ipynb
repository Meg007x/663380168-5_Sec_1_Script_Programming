# **Week4 Recap**

## 🎯 หัวข้อหลักที่คุณจะได้เรียนรู้ (What You Will Learn)

บทเรียนสัปดาห์นี้จะพาคุณก้าวข้ามขีดจำกัดของการเก็บข้อมูลในตัวแปรเดี่ยว ๆ ไปสู่ **"โครงสร้างข้อมูลแบบกลุ่ม (Collections)"** โดยเน้นเปรียบเทียบระหว่างสองโครงสร้างหลัก:

### 1. ลิสต์ (Lists - `[]`) : สายยืดหยุ่น ปรับแต่งได้ตามใจ

* **นิยาม:** โครงสร้างข้อมูลแบบเรียงลำดับ (Ordered Sequence) ที่สามารถเปลี่ยนแปลงแก้ไข เพิ่ม หรือลดข้อมูลข้างในได้ (Mutable) และอนุญาตให้มีข้อมูลซ้ำกันได้


* **การจัดการข้อมูลพื้นฐาน:**
* **การเข้าถึงข้อมูล (Accessing):** ใช้ Indexing (เริ่มต้นจาก `0`) และการหั่นลิสต์ (Slicing เช่น `list[start:end:step]`)


* **การเพิ่มข้อมูล:** ใช้ `.append()` (ต่อท้าย), `.insert()` (แทรกตำแหน่งเฉพาะ), และ `.extend()` (รวมลิสต์อื่นเข้ามาต่อท้าย)


* **การลบข้อมูล:** ใช้ `.remove()` (ลบด้วยค่า), `.pop()` (ดึงข้อมูลออกด้วยอินเด็กซ์และส่งคืนค่า), และ `del` (ลบโดยตรง)




* **เมธอดและฟังก์ชันที่เกี่ยวข้อง:**
* การตรวจสอบจำนวนด้วย `len()`

* การเรียงลำดับด้วย `.sort()` (ปรับในตัวแปรเดิม) หรือ `sorted()` (สร้างลิสต์เรียงลำดับอันใหม่)


* การกลับด้านข้อมูลด้วย `.reverse()`

* การค้นหาอินเด็กซ์ด้วย `.index()` และนับจำนวนค่าซ้ำด้วย `.count()`




### 2. ทูเพิล (Tuples - `()`) : สายมั่นคง แข็งแกร่ง ห้ามแก้ไข

* **นิยาม:** โครงสร้างข้อมูลแบบเรียงลำดับที่มีคุณสมบัติ **ห้ามแก้ไขเปลี่ยนแปลงเด็ดขาด (Immutable)**

* **การสร้างและข้อควรระวัง:**
* การสร้างทูเพิลที่มีข้อมูลเพียง 1 ตัว ต้องมีเครื่องหมายจุลภาคต่อท้ายเสมอ เช่น `single_item = ("apple",)`

* การแพ็คข้อมูลแบบไม่ใส่เล็บ (Tuple Packing) เช่น `x, y, z = 1, 2, 3`



* **ทำไมต้องใช้ Tuple?:**
* เพื่อความปลอดภัยของข้อมูลที่ห้ามแก้ไขในระบบ (เช่น พิกัดแผนที่, ค่าตั้งต้นของระบบ)


* มีความปลอดภัยและประหยัดหน่วยความจำของคอมพิวเตอร์มากกว่าลิสต์สำหรับการเก็บข้อมูลขนาดใหญ่ที่เป็นค่าคงที่
