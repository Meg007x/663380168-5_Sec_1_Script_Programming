# **Lab 4.1: Student Management System (List Operations)**

```python
# lab4_1.py - Student Management System (List Operations)
# หมายเหตุ: โค้ดนี้ตัดส่วน "Challenge" (ระบบเมนูแบบ while loop) ออก
#          เพื่อให้นักศึกษานำ concept ด้านล่างไปต่อยอดคิดเองเป็นแบบฝึกหัด

print("--- Lab 4.1: Student Management System ---")

# 1) เริ่มต้นด้วย list ว่างสำหรับเก็บชื่อนักศึกษา
student_names = []

# 2) เพิ่มนักศึกษา: รับชื่อจากผู้ใช้ 3 คน แล้ว append() เข้า list
print("\n[1] Add Students")
for i in range(3): #กำหนดตัวแปรใน loop 👈ต้องทำ
    name_input = input(f"Enter student name #{i + 1}: ") #กำหนดตัวแปร 👈ต้องทำ
    student_names.append(name_input)

print(f"Current list: {student_names}")

# 3) ค้นหานักศึกษา: ใช้ 'in' เพื่อเช็คว่ามีชื่อนี้อยู่ใน list หรือไม่
#    ถ้าพบ ให้แสดง index ด้วย index()
print("\n[2] Find a Student")
search_name = input("Enter a name to search for: ")

if search_name in student_names: #เขียนเงื่อนไข ถ้าหากมีชื่อในลิส: 👈ต้องทำ
    position = student_names.index(search_name)
    print(f"Found '{search_name}' at index {position}")
else:
    print(f"'{search_name}' not found in the list")

# 4) ลบนักศึกษา: เช็คก่อนว่ามีอยู่จริงหรือไม่ (กัน error) แล้วค่อย remove()
print("\n[3] Remove a Student")
remove_name = input("Enter a name to remove: ")

if remove_name in student_names:
    student_names.remove(remove_name) #เติมฟังก์ชัน remove(remove_name) 👈ต้องทำ
    print(f"'{remove_name}' has been removed")
else:
    print(f"'{remove_name}' not found, nothing removed")

print(f"Updated list: {student_names}")

# 5) เรียงลำดับรายชื่อตามตัวอักษรด้วย sort()
print("\n[4] Sort Students")
student_names.sort() #เติมฟังก์ชันsort()
print(f"Sorted list: {student_names}")

# 6) นับจำนวนนักศึกษาทั้งหมดด้วย len()
print("\n[5] Count Students")
print(f"Total students: {len(student_names)}") #ใช้คำสั่งlen() หาจำนวนนักศึกษาในลิส student_names👈ต้องทำ

# -----------------------------------------------------------'---------
# ให้นักศึกษาลองต่อยอดเอง (Challenge):
# ลองเขียนระบบเมนูด้วย while loop (จาก Week 3) ที่ให้ผู้ใช้เลือกได้ว่า
# จะเพิ่ม / ลบ / ค้นหา / เรียงลำดับ / หรือออกจากโปรแกรม
# โดยนำโค้ดแต่ละส่วนด้านบนไปจัดเป็นฟังก์ชันหรือเงื่อนไขในเมนู
# --------------------------------------------------------------------
```

# **Challenge 1**

```python
print("--- Challenge 1 : While loop ---")
student_names = []

while True:
  print(f"ทางเลือกพิมพ์ 1 =  เพิ่มชื่อ / 2 = ลบชื่อ / 3 = ค้นหารายชื่อ / 4 = เรียงลำดับ / 5 = ออกจากโปรแกรม / เลขอื่นก็ออกจากโปรแกรม")
  choice = input("Select option: ")


  if choice == "1":
    print("\n[1] Add Students")
    name_input = input("กรุณาเพิ่มชื่อที่ต้องการ : ")
    student_names.append(name_input)
    print(f"Current list: {student_names}")
    continue

  elif choice == "2":
    print("\n[2] Remove a Student")
    remove_name = input("Enter a name to remove: ")

    if remove_name in student_names:
      student_names.remove(remove_name)
      print(f"'{remove_name}' has been removed")
    else:
      print(f"'{remove_name}' not found, nothing removed")

    print(f"Updated list: {student_names}")
    continue

  elif choice == "3":
      print("\n[3] Find a Student")
      search_name = input("Enter a name to search for: ")
      if search_name in student_names: #เขียนเงื่อนไข ถ้าหากมีชื่อในลิส: 👈ต้องทำ
        position = student_names.index(search_name)
        print(f"Found '{search_name}' at index {position}")
      else:
        print(f"'{search_name}' not found in the list")
      continue

  elif choice == "4":
    print("\n[4] Sort Students")
    student_names.sort()
    print(f"Sorted list: {student_names}")

  else:
    print("[\n5] ออกจากโปรแกรม")

    break
```

# **Lab 4.2: Geometric Calculations (Tuples & Immutability)**

```python
# lab4_2.py - Geometric Calculations (Tuples & Immutability)
# หมายเหตุ: โค้ดนี้ตัดส่วน "Challenge" (รับพิกัดจากผู้ใช้เพื่อคำนวณระยะทาง) ออก
#          เพื่อให้นักศึกษานำ concept ด้านล่างไปต่อยอดคิดเองเป็นแบบฝึกหัด

import math

print("--- Lab 4.2: Geometric Calculations ---")

# 1) กำหนดจุดสองจุดด้วย tuple: point1 และ point2
point1 = (3 ,  4) #เติมพิกัด x y ของจุด 👈ต้องทำ
point2 = (6, 8)  #เติมพิกัด x y ของจุด 👈ต้องทำ

# 2) เข้าถึงค่าพิกัด x, y ของแต่ละจุด
print("\n[1] Access Coordinates")
print(f"Point 1: x = {point1[0]}, y = {point1[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ
print(f"Point 2: x = {point2[0]}, y = {point2[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ

# 3) ทดลองแก้ไขค่าใน tuple เพื่อแสดงให้เห็นว่า tuple เปลี่ยนแปลงไม่ได้ (immutable)
print("\n[2] Attempt Modification (Immutability Demonstration)")
try:
    point1[0] = 5  # บรรทัดนี้จะทำให้เกิด TypeError เพราะ tuple แก้ไขไม่ได้
except TypeError as e:
    print(f"Error trying to modify tuple: {e}")

# 4) คำนวณระยะห่างระหว่างจุดสองจุดด้วยสูตรระยะทาง
#    distance = sqrt( (x2 - x1)^2 + (y2 - y1)^2 )
print("\n[3] Calculate Distance")
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2) #เติมเลขตำแหน่งในลิส 0 และ 1 ใน [] ให้ถูกต้อง 👈ต้องทำ
print(f"Distance between point1 {point1} and point2 {point2}: {distance}")

# --------------------------------------------------------------------
# ให้นักศึกษาลองต่อยอดเอง (Challenge):
# ลองแก้โค้ดให้รับพิกัดของจุดใหม่ 2 จุดจากผู้ใช้ผ่าน input()
# (อย่าลืมแปลงชนิดข้อมูลจาก input() ให้เป็นตัวเลขก่อนนำไปคำนวณ)
# แล้วคำนวณระยะห่างระหว่างจุดทั้งสองที่ผู้ใช้กรอกเข้ามา
# --------------------------------------------------------------------
```

# **Activity 4.1: List vs. Tuple: Scenario Matching**

## **[1] A collection of daily temperatures recorded over a week (อุณหภูมิเฉลี่ยรายวันในรอบสัปดาห์)**

**คำตอบ** List เพราะ เนื่องจากอุณหภูมิเป็นข้อมูลที่ต้องมีการอัปเดตแบบเรียลไทม์ หรืออาจต้องมีการเพิ่มข้อมูลวันถัดไปเข้าไปในระบบเรื่อย ๆ จึงต้องการโครงสร้างที่ยืดหยุ่นและแก้ไขได้ (Mutable)

## **[2] The RGB color values for a specific shade (ค่าสี RGB ของเฉดสีเฉพาะ เช่น (255, 0, 0))**

**คำตอบ** Tuple เเพราะ ค่าสี RGB ของเฉดสีเฉพาะนั้นเป็นค่าคงที่สากลที่ไม่ควรถูกเปลี่ยนแปลงแก้ไขในภายหลัง และการใช้ Tuple จะช่วยให้ระบบประมวลผลได้รวดเร็วและปลอดภัยขึ้น (Immutable)

## **[3] Items in a shopping cart that can be added or removed (สินค้าในตะกร้าช้อปปิ้งที่เพิ่มหรือลบได้)**

**คำตอบ** List เพราะ ลูกค้าจำเป็นต้องเพิ่มสินค้าใหม่เข้าตะกร้า `(.append())` หรือลบสินค้าที่ไม่ต้องการออก `(.remove())` ได้ตลอดเวลา โครงสร้างจึงต้องเปลี่ยนแปลงได้

## **[4] The days of the week in order (รายชื่อวันในสัปดาห์ตามลำดับ)**

**คำตอบ** Tuple เพราะ วันในสัปดาห์ (จันทร์-อาทิตย์) มีจำนวนคงที่ 7 วันเสมอ ไม่มีการเพิ่มวันแปดหรือลบวันพุธทิ้ง การใช้ Tuple จึงปลอดภัยจากการแก้ไขตรรกะของระบบโดยไม่ได้ตั้งใจ

## **[5] Student scores on a series of assignments, which might be updated (คะแนนเก็บของนักเรียนที่อาจต้องมีการอัปเดต)**

**คำตอบ** List เพราะ คะแนนสอบของนักศึกษามีโอกาสปรับเปลี่ยนได้ (เช่น ส่งงานช้าแล้วโดนหักคะแนน หรืออาจารย์ปรับคะแนนเพิ่มให้) ข้อมูลจึงต้องสามารถแก้ไขรายตัวได้

## **[6] GPS coordinates for a fixed landmark (พิกัด GPS ของสถานที่สำคัญทางประวัติศาสตร์ที่ไม่มีการเคลื่อนย้าย)**

**คำตอบ** Tuple เพราะ พิกัดละติจูดและลองจิจูดของสถานที่ถาวรไม่มีการเปลี่ยนแปลง การใช้ Tuple ช่วยรับประกันว่าพิกัดนั้นจะปลอดภัยจากการแฮกหรือการเขียนโค้ดทับข้อมูลเดิม

# **Activity 4.2: Interactive To-Do List (using List Operations)**

```python
print("--- Activity 4.2: Interactive To-Do List ---")


todo_list = []

while True:
    print("\n===== MENU =====")
    print("1. Add a task (เพิ่มงาน)")
    print("2. View tasks (ดูงานทั้งหมด)")
    print("3. Mark task as complete (ลบงานที่ทำเสร็จ)")
    print("4. Exit (ออกจากโปรแกรม)")


    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        task = input("Enter the task: ")
        todo_list.append(task)
        print(f"Task '{task}' added successfully!")

    elif choice == "2":
        print("\n--- Your To-Do List ---")
        if len(todo_list) == 0:
            print("Your list is empty! No tasks to do.")
        else:

            for i in range(len(todo_list)):
                print(f"{i + 1}. {todo_list[i]}")

    elif choice == "3":
        if len(todo_list) == 0:
            print("No tasks to complete!")
        else:
            print("\n--- Current Tasks ---")
            for i in range(len(todo_list)):
                print(f"{i + 1}. {todo_list[i]}")


            try:
                task_num = int(input("Enter task number to complete: "))

                if 1 <= task_num <= len(todo_list):
                    completed_task = todo_list.pop(task_num - 1)
                    print(f"Completed and removed: '{completed_task}'")
                else:
                    print("Invalid task number!")
            except ValueError:
                print("Please enter a valid number.")

    elif choice == "4":
        print("Goodbye! Have a productive day!")
        break

    else:
        print("Invalid choice! Please select 1-4.")
```
