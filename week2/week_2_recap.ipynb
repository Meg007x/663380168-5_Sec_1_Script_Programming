#### 1. Control Flow & Conditionals (เงื่อนไขและการตัดสินใจ)

> **เปรียบเทียบ:** ลองนึกถึงเวลาคุณบุกฐานศัตรู คุณมีกฎในหัวว่า *"**ถ้า (if)** ศัตรูมีปืน ฉันจะหาที่กำบัง, **แต่ถ้า (elif)** ศัตรูไม่มีอาวุธ ฉันจะเข้าจับกุม, **นอกเหนือจากนั้น (else)** ฉันจะเฝ้าระวัง"* ใน Python เราใช้ `if`, `elif`, `else` เพื่อให้โปรแกรมรู้จักคิด แต่นี่คือจุดที่ **AI Ethics (จริยธรรม AI)** สำคัญที่สุด! เงื่อนไข (Condition) ที่คุณเขียนต้องไม่มีอคติ (Bias) เช่น ไม่ตัดสินใจปฏิเสธคนเพียงเพราะเพศหรือเชื้อชาติ

```
import logging

# ตั้งค่าสมุดบันทึกภารกิจ (Explainable AI - เพื่อให้เรารู้ว่าทำไม AI ถึงตัดสินใจแบบนี้)
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def check_security_clearance(age, has_badge):
    # ใช้ Logical Operators (and/or/not)
    # เปรียบเหมือนระบบรักษาความปลอดภัย 2 ชั้นของ MI6 (ต้องอายุถึงเกณฑ์ "และ" มีบัตร)
    if age >= 18 and has_badge:
        logging.info("Access Granted: อนุญาตให้เข้าห้องนิรภัยได้")
    elif age >= 18 and not has_badge:
        logging.warning("Access Denied: อายุถึงเกณฑ์ แต่ไม่มีบัตรผ่าน!")
    else:
        logging.error("Security Breach: ตรวจพบผู้บุกรุกอายุไม่ถึงเกณฑ์!")

# จำลองสถานการณ์
check_security_clearance(25, True)
check_security_clearance(20, False)
```

#### 2. Relational & Logical Operators (ตรรกะสายลับ)

> **เปรียบเทียบ:** นี่คือเครื่องจับเท็จของคุณ

* `==` (เท่ากับ), `!=` (ไม่เท่ากับ), `>`, `<` ใช้เทียบข้อมูล
* `and` (ต้องจริงทั้งคู่), `or` (จริงแค่อย่างใดอย่างหนึ่งก็พอ), `not` (กลับดำเป็นขาว)

ในการสร้าง **Responsible AI** เราใช้ตัวดำเนินการเหล่านี้เพื่อสร้าง "ขอบเขตความปลอดภัย (Safeguards)" ไม่ให้ AI ทำงานนอกเหนือคำสั่งที่กำหนดไว้

#### 3. List Comprehensions (สายพานการผลิตของแผนก Q)

> **เปรียบเทียบ:** สมมติคุณมีรายชื่อผู้ต้องสงสัย 100 คน และต้องการคัดเฉพาะคนที่อายุมากกว่า 30 ปี ถ้าใช้วิธีปกติ (For-loop) เหมือนคุณหยิบแฟ้มมาดูทีละอัน... มันช้า! List Comprehension เปรียบเหมือนเครื่องจักรคัดกรองอัตโนมัติของแผนก Q ที่กรองแฟ้มทั้งหมดเสร็จในพริบตาเดียว!

```
# รายชื่ออายุของผู้ต้องสงสัย (List พื้นฐาน)
suspects_ages = [15, 34, 42, 18, 29, 12, 50]

# ใช้ List Comprehension คัดกรองเป้าหมายที่อายุมากกว่า 30 ปี
# โครงสร้าง: [ผลลัพธ์ที่จะเอา for ตัวแปร in ข้อมูลทั้งหมด if เงื่อนไข]
priority_targets = [age for age in suspects_ages if age > 30]

# บันทึกผลลัพธ์ลง Log เพื่อความโปร่งใส (XAI)
logging.info(f"พบเป้าหมายสำคัญ (อายุ > 30) จำนวน {len(priority_targets)} คน: {priority_targets}")
```

### 🕵️‍♂️ รายงานสรุปและคำสั่งต่อไปจาก M:

จำไว้ว่า การเขียน `if-else` คือการสร้างสมองให้ AI และการใช้ `Logging` ควบคู่ไปกับการตัดสินใจเหล่านั้น คือการทำ **Explainable AI (XAI)** ทุกครั้งที่ AI ของคุณเลือกทางซ้ายแทนทางขวา มันต้องทิ้งร่องรอยใน Log ไว้ให้ฉันและคุณตรวจสอบได้เสมอว่า "ทำไม"
