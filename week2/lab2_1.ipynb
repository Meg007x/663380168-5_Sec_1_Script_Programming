```python
#เริ่มจากให้ผู้ใช้กรอกตัวเลขที่ต้องการเข้ามา
number = int(input("กรุณาใส่ตัวเลขที่ต้องการ : "))

#ตรวจว่าเป็น จำนวนเต็ม ใด
if number > 0: #ถ้ามากกว่า 0 ถือว่าเป็นจำนวนเต็มบวก result_type จะมีเก็บค่า "จำนวนเต็มบวก"
    result_type = "จำนวนเต็มบวก"
elif number < 0: #ถ้ามากกว่า 0 ถือว่าเป็นจำนวนเต็มลบ result_type จะมีเก็บค่า "จำนวนเต็มลบ"
    result_type = "จำนวนเต็มลบ"
else: #ถ้าไม่เป็น บวก หรือ ลบ ก็ต้องเป็นจำนวนเต็มศูนย์ result_type จะมีเก็บค่า "จำนวนเต็มศูนย์"
    result_type = "จำนวนเต็มศูนย์"

#ตรวจว่าเป็นเลขคู่หรือคี่

if number % 2 == 0:  #ถ้า number หาร 2 แล้วเหลือ 0 ถือว่าเป็นเลขคู่ result_parity จะเก็บค่า "เลขคู่"
    result_parity = "เลขคู่"
else: #ถ้าหารแล้วไม้เป็น 0 ก็ ถือว่าเป็นเลขคี่ result_parity จะเก็บค่า "เลขคี่"
    result_parity = "เลขคี่"

#Challenge
#เอามารวมกันด้วยการใช้ print(f"") แล้วใส่ตัวแปรไว้ข้างใน
print(f"เลข {number} คือ {result_type} และเป็น {result_parity}")
#สิ่งที่จะแสดงคือ เลข {ตัวเลขที่กรอก} คือ {จำนวนเต็มใดๆ} และเป็น {เลขคู่หรือคี่}
```
