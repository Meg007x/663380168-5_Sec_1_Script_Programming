**เกม** ผู้กล้าฮิมเมลผจญภัยสลัมคลองเตย


```python
# --- เริ่มต้นเกม ผู้กล้าฮิมเมลผจญภัยสลัมคลองเตย ---
print("--- ยินดีต้อนรับสู่เกม: Choose Your Own Adventure (Himmel Edition) ---")
print("คุณคือ 'ผู้กล้าฮิมเมล' ผู้รักสันติภาพและความถูกต้อง")
print("ขณะนี้คุณกำลังเดินสำรวจสภาพความเป็นอยู่ ณ สลัมคลองเตย...")
print("ทันใดนั้น! มีเด็กติดยา 2 คนถือมีดวิ่งตรงมาทางคุณเพื่อจะแทงคุณ!")

# จุดตัดสินใจที่ 1
choice1 = input("คุณจะทำอย่างไร? (พิมพ์ 1: หยุดเด็กพวกนั้นไว้ / พิมพ์ 2: ไม่ทำอะไรเลย): ")

if choice1 == "1":
    # ผลลัพธ์ทางเลือกที่ 1 (ไปต่อ)
    print("\n[สำเร็จ!] คุณใช้ทักษะผู้กล้าหยุดเด็กติดยาทั้งสองคนไว้ได้โดยไม่มีใครบาดเจ็บ")
    print("คุณนำตัวพวกเขาส่งเจ้าหน้าที่ตำรวจ แต่ในระหว่างทาง เด็กทั้งสองเริ่มทะเลาะกัน")
    print("เด็ก A: 'เพราะแกนั่นแหละขี้ขลาด เราเลยโดนจับ!'")
    print("เด็ก B: 'แกนั่นแหละที่วิ่งช้า!'")
    print("พวกเขามัวแต่โทษกันไปมา คุณจึงต้องตัดสินใจพูดอะไรบางอย่าง...")
   # จุดตัดสินใจที่ 2 (Nested Conditional - เงื่อนไขซ้อน)
    choice2 = input("คุณจะพูดว่าอะไร? (พิมพ์ A: ปลอบใจ / พิมพ์ B: สั่งสอน): ").upper()

    if choice2 == "A":
        print("\nคุณตอบว่า: 'ไม่เป็นไรนะชีวิตเราเริ่มใหม่ได้ กลับตัวกลับใจซะ อย่าไปยุ่งกับของพวกนี้อีก'")
        print("เด็กๆ เงียบไปและถูกตำรวจคุมตัวไปเพื่อบำบัด... เกมจบลงด้วยความหวัง")

    elif choice2 == "B":
        # ผลลัพธ์ที่ดีที่สุดตามเนื้อเรื่องของผู้ใช้
        print("\nคุณตอบว่า: 'พวกแกทั้งสองคนนั่นแหละที่ผิด! เลิกโทษคนอื่นแล้วไปสำนึกผิดในคุกซะ!'")
        print("--------------------------------------------------")
        print("!!! ยินดีด้วย คุณได้รับฉากจบที่ดีที่สุด (TRUE ENDING) !!!")
        print("นี่คือสิ่งที่ผู้กล้าฮิมเมลควรจะทำ... ส่งเด็กเช้าคุก")
        print("--------------------------------------------------")

    else:
        print("\nคุณนิ่งเงียบไม่พูดอะไร... เด็กๆ ถูกจับไปแบบงงๆ")

elif choice1 == "2":
    # ผลลัพธ์ทางเลือกที่ 2 (Game Over)
    print("\n[พลาดแล้ว!] คุณเลือกที่จะไม่ทำอะไรเลย...")
    print("เด็กติดยาแทงเข้าที่สีข้างของคุณ")
    print("[คุณตายแล้ว GGEZ]")
```
