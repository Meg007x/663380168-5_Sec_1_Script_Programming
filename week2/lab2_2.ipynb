```python
#เริ่มจากรนับอายุของผู้ใช้
age = int(input("กรุณาใส่อายุของคุณ : "))

#ตรวจสอบอายุ
if age < 5: #ถ้าอายุน้อยกว่า 5 ปี ไม่ให้เข้า
   print("คุณเด็กเกินไปสำหรับโรงหนัง! ไปดูการ์ตูนเถอะ") #ผลลัพธ์ คุณเด็กเกินไปสำหรับโรงหนัง! ไปดูการ์ตูนเถอะ
elif age <= 12: #ถ้าอายุน้อยกว่าหรือเท่ากับ 12 ปี จะแนะนำเรตหนังตามเงื่อนไข
  print("แนะนำหนังเรท G หรือ PG") #ผลลัพธ์ แนะนำหนังเรท G หรือ PG
elif age <= 17: #ถ้าอายุน้อยกว่าหรือเท่ากับ 17 ปี จะแนะนำเรตหนัง PG-13
  print("แนะนำหนังเรท PG-13") #ผลลัพธ์ แนะนำหนังเรท PG-13
else: #ในกรณีที่อายุ 18 ปีขึ้นไป
    print("คุณดูหนังได้ทุกเรท") #ผลลัพธ์ คุณดูหนังได้ทุกเรท

    # ถามว่าชอบหนังแอคชั่นมั้ย โดยรับค่ามาจากผู้ใช้ตอบเป็น yes/no บังคับเป็นตัวเล็ก
    choice = input("คุณชอบหนังแอคชั่นไหม? (yes/no): ").lower()

    #CHALLENGE
    #ในเงื่อนไขที่ว่า อายุมากว่า 18 ถือว่าเป็น True และ choice = yes จะมีค่าเป็น True
    #True and True จะได้ True
    if age >= 18 and choice == "yes":
         print("งั้นเราขอแนะนำหนังแอคชั่นบล็อกบัสเตอร์เรื่องล่าสุดให้คุณด้วย!") #ผลลัพธ์ งั้นเราขอแนะนำหนังแอคชั่นบล็อกบัสเตอร์เรื่องล่าสุดให้คุณด้วย!
    #ในกรณีที่ไม่ต้องการดูหนังแนวแอคชั่น
    else:
        print("ไม่เป็นไร เรายังมีหนังแนวอื่นๆ แนะนำคุณอีกนะ") #ผลลัพธ์ ไม่เป็นไร เรายังมีหนังแนวอื่นๆ แนะนำคุณอีกนะ
```
