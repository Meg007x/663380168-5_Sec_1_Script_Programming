# 1. การตั้งค่าระบบบันทึก (Setting up the Secret Log)
**เปรียบเทียบ:** แทนที่จะให้ AI พูดพล่ามทุกอย่างออกหน้าจอ เราจะตั้งค่าสมุดปกดำให้บันทึกเฉพาะเรื่องสำคัญ โดยมีการระบุเวลาและระดับความรุนแรงของสถานการณ์ (Severity Levels)

```python
import logging

# 1. ตั้งค่าพื้นฐานสำหรับสมุดบันทึก (Responsible & Explainable Setup)
logging.basicConfig(
    level=logging.INFO, # กำหนดระดับขั้นต่ำที่จะบันทึก (INFO ขึ้นไป)
    format='%(asctime)s - %(levelname)s - %(message)s' # รูปแบบ: วันเวลา - ระดับ - ข้อความ
)

logging.info("ภารกิจสัปดาห์ที่ 1 เริ่มต้นขึ้นแล้ว... ระบบ Logging พร้อมใช้งาน")
```

# 2. ตัวแปรและประเภทข้อมูล (Variables & Data Types)
**เปรียบเทียบ:** ข้อมูลแต่ละประเภทเปรียบเหมือนอุปกรณ์ในกระเป๋าของสายลับ ของแต่ละชิ้นต้องใส่ให้ถูกช่องและห้ามปนกัน ตัวแปร (Variable) คือป้ายแปะหน้ากล่องใส่อุปกรณ์เหล่านั้น

```python
# ป้อนข้อมูลพื้นฐาน (อ้างอิงจากบทเรียนเรื่อง Data Types)
agent_name = "James"          # ข้อความ (String - str) -> ชื่อสายลับ
agent_id = 7                  # เลขจำนวนเต็ม (Integer - int) -> รหัสลับ
mission_success_rate = 99.5    # เลขทศนิยม (Float - float) -> อัตราสำเร็จ
is_active = True              # ค่าความจริง (Boolean - bool) -> สถานะปฏิบัติการ

# บันทึกข้อมูลเข้าสมุดอนุทิน (ใช้ f-string ในการจัดฟอร์แมตข้อความ)
logging.info(f"เจ้าหน้าที่: {agent_name} | รหัส: {agent_id:03d}")
logging.info(f"ประเภทข้อมูลของ agent_id คือ {type(agent_id)}")
```

# 3. การรับข้อมูลจากผู้ใช้และการแปลงประเภทข้อมูล (Interactive Input & Type Casting)
**เปรียบเทียบ:** เมื่อคุณสอบปากคำพยาน ข้อมูลที่ได้จะเป็นคำพูด (String) เสมอ แต่ถ้าพยานบอกตัวเลข เช่น "อายุ 25 ปี" แล้วคุณต้องการเอาไปคำนวณปีเกิด คุณต้องแปลงคำพูดนั้นให้เป็นตัวเลขจริงๆ ก่อน (เรียกว่า Type Casting)

```python
# จำลองการใช้ฟังก์ชัน input() จากบทเรียน (แต่ในระบบอัตโนมัติจะแปลงค่าทันที)
# สมมติผู้ใช้พิมพ์ตัวเลขเข้ามาทางคีย์บอร์ด โค้ดจะได้รับเป็น "2026" (String)
input_year_str = "2026"

# การแปลงประเภทข้อมูล (Type Casting) จาก str เป็น int เพื่อความปลอดภัยของระบบ (Responsible AI)
current_year = int(input_year_str)
next_year = current_year + 1

logging.info(f"ปีปัจจุบันของภารกิจ: {current_year} (ตรวจสอบประเภทแล้ว: {type(current_year)})")
logging.warning(f"คำเตือน: กำลังจะก้าวเข้าสู่ปีภารกิจถัดไปคือ {next_year}!")
```

## 🕵️‍♂️ ทำไมภารกิจนี้ถึงสอดคล้องกับหลักการของ M?

**1.Explainable AI (อธิบายได้):** การใช้ ***logging.basicConfig(format='...')*** ทำให้เวลาโค้ดเกิดปัญหา มันจะแจ้งวันเวลาและลำดับเหตุการณ์ชัดเจน ทำให้คุณเดินมาอธิบายฉันได้ว่า โค้ดมันผิดพลาดที่บรรทัดไหนและเพราะอะไร ไม่ใช่ปล่อยให้เป็นกล่องดำ

**2.Responsible AI (รับผิดชอบและปลอดภัย):** การใช้ ***logging.warning() หรือ logging.error()*** ช่วยแยกแยะระดับความสำคัญของข้อมูล ทำให้ระบบรักษาความปลอดภัยของ MI6 ตรวจจับสิ่งผิดปกติ (Anomaly) ได้ทันท่วงทีก่อนที่ระบบจะพังพินาศ
