### 🕵️‍♂️ Part 1: ตรรกะและการตัดสินใจ (Control Flow)

* **เครื่องหมายเปรียบเทียบ (Relational):** `==` (เท่ากันไหม?), `!=` (ไม่เท่ากันใช่ไหม?), `>`, `<`, `>=`, `<=` ทุกตัวส่งผลลัพธ์กลับมาเป็นค่า Boolean แค่สองอย่างคือ **`True`** หรือ **`False`**
* **ตารางตรรกศาสตร์ (Logical):**
* `and` = ต้องจริง **ทั้งคู่** ถึงจะยอมปล่อยผ่าน

* `not` = สลับขั้วความจริง (กลับดำเป็นขาว)


* **ทางเลือกเดินรถ (`if-elif-else`):** คอมพิวเตอร์จะเช็กเงื่อนไขจากบนลงล่าง ถ้าเจออันไหนที่เป็น `True` จะวิ่งเข้าห้องนั้นทันที และ **ห้ามลืมกด Tab ย่อหน้า (Indentation)** ให้ลูกน้องเด็ดขาด!
* **เงื่อนไขซ้อนเงื่อนไข (Nested `if`):** การตั้งคำถามซ้อนเป็นชั้นๆ เหมือนกล่องเซฟซ้อนกล่องเซฟ ด่านแรกต้องผ่านก่อน ด่านในสุดถึงจะทำงาน

# **🚀 Week 3 Snapshot  5 mins**

### 🕵️‍♂️ 1. การทำซ้ำพื้นฐาน (for & while Loops)

* **`for` Loop (รู้จำนวนรอบแน่นอน):** ใช้ท่องไปในข้อมูลที่เป็นลำดับ (Sequence) หรือใช้คู่กับ `range()`
* `range(5)` $\rightarrow$ 0, 1, 2, 3, 4 (ไม่รวมตัวท้าย)
* `range(1, 10, 2)` $\rightarrow$ เริ่ม 1, ถึงก่อน 10, กระโดดทีละ 2 (1, 3, 5, 7, 9)


* **`while` Loop (ไม่รู้จำนวนรอบที่แน่ชัด):** จะทำซ้ำไปเรื่อยๆ **ตราบใดที่เงื่อนไขยังคงเป็น True** (ระวัง! ต้องมีบรรทัดอัปเดตค่าเสมอเพื่อป้องกันลูปนรกไม่สิ้นสุด หรือ Infinite Loop)

### 🚨 2. คำสั่งควบคุมสายพาน (Loop Control)

* **`break`:** สั่งให้ระเบิดลูปทิ้งแล้วเดินออกจากลูปทันที (มักใช้กับการดักจับภัยคุกคาม)
* **`continue`:** สั่งให้ข้ามโค้ดที่เหลือในรอบนั้น แล้ววาร์ปขึ้นไปเริ่มรอบถัดไปทันที

### 🚀 3. เทคนิคขั้นสูงของแผนก Q (Nested Loops & Generators)

* **Nested Loops:** ลูปซ้อนลูป (ลูปนอกเดิน 1 รอบ ลูปในต้องเดินจนครบ) มักใช้สร้างตาราง วาดแพทเทิร์นดาว หรือประมวลผลเมทริกซ์ข้อมูล
* **Generator Functions (`yield`):** ฟังก์ชันระดับสูงที่เปลี่ยนจากคำสั่ง `return` มาเป็น **`yield`** มันจะส่งค่าออกมาทีละตัวเมื่อถูกเรียกใช้ (`next()`) ไม่จองแรมทั้งหมดพร้อมกัน เหมาะสำหรับการประมวลผลข้อมูลอนันต์ (Infinite Series) หรือข้อมูลขนาดใหญ่เพื่อความปลอดภัยของระบบ (**Responsible AI**)

### 💻 Code Snippets สำหรับ Learning Logging (XAI & Ethics Included)

นี่คือโค้ดตัวอย่างที่รวมเนื้อหาหลักของ Week 3 พร้อมติดระบบกล้องวงจรปิด `logging` เพื่อความโปร่งใสตรวจสอบได้ครับ:

```python
import logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# --- 1. การใช้ while loop และ break/continue (ระบบสแกนวัตถุต้องสงสัย) ---
logging.info("=== เริ่มภารกิจสแกนกระเป๋าด้วยระบบ Loop Control ===")
cargo_weights = [12, 5, -3, 45, 999, 15] # -3 คือขยะ, 999 คือระเบิด (Edge Cases)
index = 0

while index < len(cargo_weights):
    weight = cargo_weights[index]
    index += 1
    
    # ดักจับข้อมูลขยะ (Responsible AI)
    if weight < 0:
        logging.warning(f"น้ำหนักติดลบ ({weight}) ข้อมูลผิดพลาด! ข้ามไปตรวจชิ้นต่อไป")
        continue # วาร์ปไปรอบถัดไป
        
    # ดักจับภัยคุกคามร้ายแรง
    if weight > 500:
        logging.error(f"🚨 ตรวจพบวัตถุอันตรายร้ายแรง น้ำหนัก {weight} กิโลกรัม! สั่งล็อกดาวน์ระบบด่วน")
        break # ระเบิดลูปทิ้งทันที เพื่อความปลอดภัย


# --- 2. การใช้ Nested Loops (เครื่องยิงสเปรย์เลเซอร์สแกนพื้นที่ 3x3) ---
logging.info("\n=== เริ่มทำงานเลเซอร์แบบสแกนพิกัด (Nested Loops) ===")
for row in range(1, 4):      # ลูปนอก: แถวที่ 1 ถึง 3
    for col in range(1, 4):  # ลูปใน: คอลัมน์ที่ 1 ถึง 3
        # พิมพ์พิกัดที่ AI กำลังประมวลผล (Explainable AI)
        print(f"[{row},{col}] ", end="")
    print() # ขึ้นบรรทัดใหม่เมื่อจบลูปในแต่ละแถว


# --- 3. Generator Functions กับรหัส Fibonacci (ดึงเนื้อหาจากสไลด์ p3) ---
logging.info("\n=== เรียกใช้งาน Generator ผลิตรหัสลับ Fibonacci ===")

def fibonacci_generator():
    """ฟังก์ชันผลิตตัวเลขแบบประหยัดแรม (Lazy Evaluation) ด้วย yield"""
    a, b = 0, 1
    while True:
        yield a # ส่งค่า a ออกไป แล้วหยุดรอตรงนี้จนกว่าจะโดนเรียกอีกครั้ง
        a, b = b, a + b

# เปิดเครื่อง Generator
fib_machine = fibonacci_generator()

# ดึงรหัสลับมาใช้ 5 ตัวแรกโดยใช้ฟังก์ชัน next()
for i in range(5):
    secret_code = next(fib_machine)
    logging.info(f"รหัสลับลำดับที่ {i+1} คือ: {secret_code}")

```

### 🕵️‍♂️ 1. คลังเก็บข้อมูลลำดับขั้นสูง (Lists vs Tuples)

* **Lists `[ ]` (คลังเก็บของที่เปลี่ยนแปลงได้ - Mutable):** ในสัปดาห์ที่ 2 และ 3 เราใช้มันเก็บข้อมูลแบบผ่านๆ แต่ใน Week 4 เราจะเจาะลึกคำสั่งจัดการไส้ในของมัน เช่น `.append()` (เพิ่มของ), `.pop()` (ดึงของออก), และ `.sort()` (จัดเรียงกำลังพล)
* **Tuples `( )` (กล่องนิรภัยล็อกตาย - Immutable):** ข้อมูลที่ห้ามใครแก้ไขเด็ดขาดหลังจากสร้างเสร็จ เช่น ค่าพิกัดฐานทัพ หรือรหัสประจำตัวของ M เพื่อความปลอดภัยขั้นสูงของระบบ (**Responsible AI**)

### 🔐 2. พจนานุกรมรหัสและเซตลบลายนิ้วมือ (Dictionaries & Sets)

* **Dictionaries `{key: value}`:** การสืบค้นข้อมูลสายลับรายบุคคลโดยไม่ต้องวนลูปหาให้เสียเวลา ค้นผ่านคำหลัก (Key) แล้วเจอข้อมูลทันที
* **Sets `{ }`:** เครื่องมือสกัดข้อมูลซ้ำระดับสูงที่เราแอบเห็นผ่านๆ ในเซตคอมเพรแฮนชัน คราวนี้เราจะเรียนการหาจุดร่วม (`.intersection()`) และการรวมฐานข้อมูล (`.union()`) ของข้อมูลสองชุด

### 💻 คาดการณ์ Code Snippets สำหรับเขียนโครงสร้างฐานข้อมูล (Learning Logging)

นี่คือแบบจำลองโค้ดโครงสร้างข้อมูลที่เรามีโอกาสสูงมากที่จะได้เจอใน Week 4 พร้อมติดตั้ง `logging` เพื่อแสดงความโปร่งใส (**Explainable AI**):

```python
import logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

logging.info("=== [คาดการณ์] จำลองการทำงานโครงสร้างข้อมูล Week 4 ===")

# --- 1. List & Tuple: การเก็บรหัสปฏิบัติการและการล็อกพิกัดฐานทัพ ---
agent_tools = ["Laser Pen", "Grappling Hook", "Smoke Bomb"]
agent_tools.append("Smart Watch") # แผนก Q ส่งอุปกรณ์มาเพิ่มใน List ได้
logging.info(f"คลังอาวุธปัจจุบัน (List เปลี่ยนแปลงได้): {agent_tools}")

# พิกัดตึก MI6 (ใช้ Tuple เพราะห้ามใครแฮกมาเปลี่ยนตำแหน่งตึกเด็ดขาด! - Responsible AI)
mi6_location = (51.4872, -0.1244)
logging.info(f"พิกัดสำนักงานใหญ่ (Tuple ล็อกตาย): {mi6_location}")


# --- 2. Dictionary: แฟ้มประวัติสายลับเพื่อการเข้าถึงความเร็วสูง ---
# ไม่ต้องวนลูปหา แค่เรียก Key ก็ดึงข้อมูลได้ทันที (Scannability & Efficiency)
agent_registry = {
    "007": {"name": "James Bond", "status": "Active"},
    "006": {"name": "Alec Trevelyan", "status": "MIA"}
}

# ดึงข้อมูลตรวจสอบย้อนกลับ (Explainable AI)
target_agent = "007"
if target_agent in agent_registry:
    logging.info(f"ดึงฐานข้อมูล {target_agent}: {agent_registry[target_agent]['name']} สถานะ: {agent_registry[target_agent]['status']}")


# --- 3. Set: การตรวจสอบรายชื่อผู้บุกรุกไม่ให้ซ้ำกัน ---
scanned_fingerprints = {"UserA", "UserB", "UserA", "UserC"} # UserA ซ้ำ
# Set จะกรองตัวซ้ำออกให้เหลืออันเดียวโดยอัตโนมัติ
logging.info(f"รายชื่อลายนิ้วมือที่แปลกปลอม (ลบตัวซ้ำแล้ว): {scanned_fingerprints}")

```

## 📖 2. บทเรียน Week 3: Control Flow: Loops (พร้อม Code Snippets)

### 📌 หัวข้อที่ 1: ความจำเป็นของลูป และการใช้งาน `for` Loop + `range()`

คอมพิวเตอร์ถูกสร้างมาเพื่อทำงานที่ซ้ำซากแทนมนุษย์ แทนที่เราจะเขียนคำสั่ง `print()` ด้วยมือ 100 ครั้ง เราสามารถสั่งให้ระบบทำงานอัตโนมัติได้ในบรรทัดเดียวด้วย `for` Loop ซึ่งจะท่องไปในข้อมูลที่เป็นลำดับ (Sequence)

#### 🛠️ อาวุธเสริม: ฟังก์ชัน `range()`

* `range(stop)`: เริ่มจาก 0 ถึงก่อนตัวหยุด


* `range(start, stop)`: เริ่มจากตัวแรก ถึงก่อนตัวหยุด


* `range(start, stop, step)`: เพิ่มค่าตามจำนวน step ที่กำหนด



```python
import logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

logging.info("--- เริ่มงาน Task 3.1: ทดสอบพลังของ for Loop ---")

# 1. วนลูปตามจำนวนรอบที่กำหนดด้วย range()
for i in range(3): # ทำงาน 3 รอบ: 0, 1, 2
    logging.info(f"รอบการทำงานอัตโนมัติที่: {i}")

# 2. การนับเลขแบบระบุจุดเริ่ม จุดจบ และสเต็ปการกระโดด
logging.info("คัดกรองเฉพาะเลขคี่จาก 1 ถึง 9:")
for odd_num in range(1, 10, 2): # เริ่ม 1, ถึงก่อน 10, เพิ่มทีละ 2
    print(f"เลขคี่: {odd_num}")

# 3. การแกะรหัสผ่านข้อความทีละตัวอักษร (Iterating over strings)
secret_msg = "MI6"
logging.info(f"เริ่มสแกนอักขระในข้อความ '{secret_msg}':")
for char in secret_msg:
    print(f"อักขระที่ตรวจพบ: {char}")
```[cite: 13]

---

### 📌 หัวข้อที่ 2: ความยืดหยุ่นของ `while` Loop และตัวควบคุม (`break` / `continue`)
`while` Loop จะทำงานซ้ำไปเรื่อยๆ **ตราบใดที่เงื่อนไขยังคงคำนวณได้ค่าเป็น True** สิ่งสำคัญที่สุดคือต้องสร้างตรรกะในลูปเพื่ออัปเดตค่าเงื่อนไขเสมอ ไม่เช่นนั้นระบบจะติดอยู่ในลูปนรกไม่สิ้นสุด (Infinite Loop)[cite: 13]

*   `break`: สั่งระเบิดและตัดการทำงานของลูปปัจจุบันทิ้งทันที[cite: 13]
*   `continue`: สั่งข้ามคำสั่งที่เหลือในรอบนั้น แล้วย้อนกลับไปเริ่มรอบถัดไปทันที[cite: 13]

```python
logging.info("\n--- เริ่มงาน Task 3.2: ระบบตรวจจับและควบคุมลูปด้วย while ---")

# 1. การใช้ while ตรวจสอบเงื่อนไข และการป้องกันลูปค้างด้วยการลดค่า (Update)
countdown = 3
while countdown > 0:
    logging.info(f"สัญญาณนับถอยหลัง: {countdown}")
    countdown -= 1 # อัปเดตค่าเพื่อไม่ให้เกิด Infinite Loop!
print("Blast off! 🚀")

# 2. การประยุกต์ใช้ break และ continue ร่วมกับการดักจับสิ่งแปลกปลอม
logging.info("เริ่มระบบกรองตัวเลขลับ (ข้ามเลขคี่, หยุดเมื่อเจอเลขต้องห้าม 6):")
for num in range(1, 10):
    if num % 2 != 0:
        # หากเป็นเลขคี่ ให้ข้ามโค้ดบรรทัดล่างไปเริ่มรอบถัดไปทันที
        continue[cite: 13]
        
    if num == 6:
        logging.warning(f"🚨 ตรวจพบเลขต้องห้ามระบบ {num}! สั่งระบบหยุดทำงาน (break)")
        break[cite: 13]
        
    print(f"เลขคู่ที่ปลอดภัย: {num}")
```[cite: 13]

---

### 📌 หัวข้อที่ 3: โครงสร้างลูปซ้อนลูป (Nested Loops)
ลูปซ้อนลูป คือการใส่ลูปโครงสร้างไว้ข้างในอีกลูปหนึ่ง ตรรกะการทำงานคือ **"ลูปภายนอกเดิน 1 รอบ ลูปภายในจะต้องทำงานจนครบทุกรอบของมัน"** เทคนิคนี้จำเป็นมากในการจัดการข้อมูลแบบตาราง ตารางพิกัด 2 มิติ หรือการวาดภาพกราฟิกด้วยอักขระ[cite: 13]

```python
logging.info("\n--- เริ่มงาน Task 3.3: การสร้าง Grid 2 มิติด้วย Nested Loops ---")

# โจทย์: สร้างตารางพิกัดจำลองขนาด 3 แถว x 4 คอลัมน์
rows = 3
cols = 4

logging.info(f"กำลังขึ้นรูปตารางขนาด {rows}x{cols}:")
for r in range(1, rows + 1):     # ลูปนอก: ควบคุมแถว (Row)
    for c in range(1, cols + 1): # ลูปใน: ควบคุมคอลัมน์ (Column)
        # คำนวณผลคูณพิกัด จัดฟอร์แมตให้กว้าง 4 ตัวอักษรเพื่อความสวยงามสมมาตร
        print(f"{r * c:4d}", end="")
    print() # จบคอลัมน์ในแถวนั้นแล้ว สั่งขึ้นบรรทัดใหม่เพื่อเริ่มแถวถัดไป
```[cite: 13]

---

### 🛡️ บันทึกท้ายคาบจากคอมมานเดอร์ M

บทเรียนชุดนี้ถูกบรรจุลงในบอร์ดคันบังของคุณเรียบร้อยแล้ว[cite: 11] หน้าที่ต่อไปของคุณคือการนำตรรกะใน Code Snippets เหล่านี้ไปลุยต่อใน **Lab 3.1 (สร้างตารางสูตรคูณ)** และ **Lab 3.2 (เกมทายตัวเลขปริศนา)** ให้บรรลุผลสำเร็จตามเกณฑ์ Rubric ที่เราตั้งไว้![cite: 11, 13]

ตอนนี้ วันที่ **8 กรกฎาคม 2026** คุณยังมีเวลาเตรียมตัวและทดสอบระบบอย่างเต็มที่ก่อนจะถึงเดดไลน์ในวันที่ **17 กรกฎาคม 2026** ครับ[cite: 11]

```
