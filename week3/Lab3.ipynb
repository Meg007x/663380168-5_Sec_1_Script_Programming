# **Lab3.1  :** โปรแกรมสร้างสูตรคูณ (Multiplication Table Generator)

**Part 1:** ตารางสูตรคูณแม่เดียว

```python
n = int(input("Enter a number for its multiplication table (e.g., 7): "))
for i in range(1, 13):
  print(f"{n} * {i} = {n * i}")
```

**Part 2:** ตารางสูตรคูณแบบเต็ม

```python
print("12x12 Multiplication Table:")
for i in range(1,13): #เติมตัวแปรให้สมบูรณ์
    for j in range(1,13): #เติมตัวแปรให้สมบูรณ์
        print(f"{i*j:4d}",end="") #คำใบ้ print ออกมาในรูปแบบ ตาราง เหมือนกันกับด้านบน แต่ต้องมีการเว้นวรรคให้เป็นตาราง
    print()
```

# **Lab 3.2:** โปรแกรมการนับถอยหลัง & เกมทายตัวเลข

**Part 1:** โปรแกรมตัวนับถอยหลัง

```python
start = int(input("กรุณาใส่ตัวเลข : "))
while start >=0:
  print(start)
  start -= 1
print("Blast off!")
```

**Part 2:** เกมทายตัวเลขอย่างง่าย

```python
secret_number = 42
attempts = 0
max_attempts = 5

print(f"--- เกมทายตัวเลข (ทายได้สูงสุด 5 ครั้ง) ---")

while attempts < max_attempts:
    guess = int(input(f"รอบที่ {attempts + 1}: ทายตัวเลข (1-100): "))
    attempts += 1

    if guess == secret_number:
        print("Congratulations! You guessed it!")
        break
    elif guess < secret_number:
        print("Too low! Try again.")
    else:
        print("Too high! Try again.")

if guess != secret_number:
    print(f"เสียใจด้วย! รอบของคุณหมดแล้ว เลขที่ถูกต้องคือ {secret_number}")
```
